# 2. Исследовательский анализ данных (EDA)
## 2.1 Описание набора данных
Я буду проводить свое исследование на наборе данных [Default of Credit Card Clients Dataset](https://www.kaggle.com/datasets/uciml/default-of-credit-card-clients-dataset) с сайта [kaggle](https://www.kaggle.com/). Он содержит данные наблюдений, собранных банком на Тайване по пользователям кредитных карт в период с апреля по сентябрь 2005 года. В качестве целевой переменной в наборе используется бинарная переменная `default.payment.next.month`. Она указывает, допустит ли владелец карты дефолт в следующем месяце. Далее в качестве объясняющих переменных используются следующие 23 признака:

* Личная информация клиента:
    * `LIMIT_BAL` - сумма предоставленного кредитного лимита (в новых тайваньских долларах - NT$).
    * `SEX` - пол: 1 - мужской, 2 - женский.
    * `EDUCATION` - образование: 1 - магистратура/аспирантура, 2 - университет, 3 - средняя школа, 4 - другое, 5 и 6 - неизвестно.
    * `MARRIAGE` - семейное положение: 1 - женат/замужем, 2 - холост/не замужем, 3 - другое.
    * `AGE` - возраст (лет).
* Информация о статусах погашения платежей:
    * `PAY_0` - статус погашения в сентябре 2005г.
    * `PAY_2` - статус погашения в августе 2005г.
    * `PAY_3` - статус погашения в июле 2005г.
    * `PAY_4` - статус погашения в июне 2005г.
    * `PAY_5` - статус погашения в мае 2005г.
    * `PAY_6` - статус погашения в апреле 2005г.

> Набор данных содержит следующие статусов: -1 - своевременная оплата, 1 - задержка платежа на один месяц, 2 - задержка платежа на два месяца, ..., 8 - задержка платежа на восемь месяцев, 9 - задержка платежа на девять и более месяцев.

* Информация о балансовых остатках:
    * `BILL_AMT1` - остаток по счету за сентябрь 2005г.
    * `BILL_AMT2` - остаток по счету за август 2005г.
    * `BILL_AMT3` - остаток по счету за июль 2005г.
    * `BILL_AMT4` - остаток по счету за июнь 2005г.
    * `BILL_AMT5` - остаток по счету за май 2005г.
    * `BILL_AMT6` - остаток по счету за апрель 2005г.

* Информация о сумме предыдущего платежа:
    * `PAY_AMT1` - сумма предыдущего платежа в сентябре 2005 г.
    * `PAY_AMT2` - сумма предыдущего платежа в августе 2005 г.
    * `PAY_AMT3` - сумма предыдущего платежа в июле 2005 г.
    * `PAY_AMT4` - сумма предыдущего платежа в июне 2005 г.
    * `PAY_AMT5` - сумма предыдущего платежа в мае 2005 г.
    * `PAY_AMT6` - сумма предыдущего платежа в апреле 2005 г.
